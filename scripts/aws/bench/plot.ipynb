```python
# Build a ping percentile plot

import matplotlib.pyplot as plt
from numpy import percentile
from json import load
import math

ROOT="/run/media/haxolotl/Backup/GitRoot/randpiper-rs"
```

```python
FILE=f"{ROOT}/data/Jan18/t1.micro/bench-processed.json"

data = {}

with open(FILE) as json_file:
    data = load(json_file)
```

```python
plt.rc('text',usetex=True)

plt.plot(data["bi_sh_gen"]["x"], data["bi_sh_gen"]["y"], label="Accumulator Share Gen", linestyle='--', marker='o')
plt.plot(data["bi_sh_vrfy"]["x"],data["bi_sh_vrfy"]["y"], label="Accumulator Share Verify", linestyle='--', marker='+')

plt.legend()
plt.xlabel("Number of Nodes ($n$)")
plt.ylabel("Computation Time (in ms)")
# plt.show()
# plt.xticks(log_points)
plt.savefig(f"{ROOT}/Plots/Bench/Bench-accumulator.pdf", bbox_inches='tight')
```
